What message will eventually appear in the sky?

```python
import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

file=open('Day-10-Data.txt',"r");
#text = file.readlines();

def fargle(x):
    y = []
    for char in x.split():
        y.append(int(char.replace(",","")))
    return y

posx = np.array([])
posy = np.array([])
velx = np.array([])
vely = np.array([])

for line in file:
    [x,y,vx,vy] = list(map(int, re.findall(r'[-\d]+',line)))
    posx = np.append(posx,x)
    posy = np.append(posy,-y)
    velx = np.append(velx,vx)
    vely = np.append(vely,-vy)

# determine smallest distance
distx = []
for i in range(0,100000):
    distx.append(max(posx + i*velx))
disty = []
for i in range(0,100000):
    disty.append(max(posy + i*vely))

# determine seconds for smallest distance
tx = distx.index(min(distx))
ty = disty.index(min(disty))
    
# setup graph
fig, ax = plt.subplots(figsize=(6,6))

# set axis ranges
ax.set_xlim(min(distx)-100,min(distx)+100) 
ax.set_ylim(min(disty)-100,min(disty)+100)

def animate(i):
    scat.set_offsets(np.c_[posx+(tx-20)*velx + i*velx, posy + (ty-20)*vely+ i*vely])
    return scat,

scat = ax.scatter(posx,posy, s=2)
ani = FuncAnimation(fig, animate, frames=50)

HTML(ani.to_jshtml())
```
